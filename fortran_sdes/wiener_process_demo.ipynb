2) `cd` into the directory containing this notebook.
3) `export CMAKE_PREFIX_PATH=<my_fortran_stdlib_path>`. If you used the Snap store to install in step 1, then you should have `<my_fortran_stdlib_path> = "usr/local/lib/cmake/fortran_stdlib"`. 
4) `mkdir build`
5) `cd build`
6) `cmake ..`
7) `cmake --build .`
8) `./baby_sde.exe`

If you get an **error** instead of just a *warning* when running steps 6-8, please open an issue!
